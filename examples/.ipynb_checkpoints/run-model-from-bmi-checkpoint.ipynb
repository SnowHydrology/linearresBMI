# Run the `linearresBMI` model using Basic Model Interface commands

This notebook demonstrates the execution of a linear reservoir model to estimate discharge using BMI. View the source code for the [model](https://github.com/SnowHydrology/linearresBMI/blob/master/linearres/linearres.py) and its [BMI](https://github.com/SnowHydrology/linearresBMI/blob/master/linearres/bmi_linearres.py) on GitHub.

This exercise is based on the [heat](https://github.com/csdms/bmi-example-python) example from [CSDMS](https://csdms.colorado.edu/wiki/Main_Page).

Start by importing `os`, `numpy`, `pandas`, `matplotlib`, and the `Linearres` BMI:

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from linearres import LinearresBmi
```

Create an instance of the model's BMI.

```python
x = LinearresBmi()
```

Use the BMI `get_component_name` function to query the model's name.

```python
print(x.get_component_name())
```

Start the `Linearres` model through its BMI using a configuration file. First, take a look at the file and then run the BMI `initialize` function.

```python
cat linearres.yaml
```

```python
x.initialize("linearres.yaml")
```

Check the time information for the model.

```python
print("Start time:", x.get_start_time())
print("End time:", x.get_end_time())
print("Current time:", x.get_current_time())
print("Time step:", x.get_time_step())
print("Time units:", x.get_time_units())
```

Show the input and output variables for the component (aside on [Standard Names](https://csdms.colorado.edu/wiki/CSDMS_Standard_Names)):

```python
print("Input vars =", x.get_input_var_names())
print("Output vars =", x.get_output_var_names())
```

For a quick example, set precipitation to a constant value using standard BMI functions.

```python
# when the model is initialized, precipitation equals 0
# we can check this using either get_value or get_value_ptr
tmp_array = np.zeros(1,)
x.get_value('atmosphere_water__precipitation_leq-volume_flux', tmp_array)
print("Precipitation from get_value =", tmp_array)
print("Precipitation from get_value_ptr =", x.get_value_ptr('atmosphere_water__precipitation_leq-volume_flux'))

# Now we can set the value of precipitation
# first make a numpy single-element array and give it a value
precip = np.full(1, 5)
x.set_value("atmosphere_water__precipitation_leq-volume_flux", precip)
print("Precipitation from get_value_ptr after set_value=", x.get_value_ptr('atmosphere_water__precipitation_leq-volume_flux'))
```

A key advantage of BMI is that **we don't have to know the names of any of the input or output variables**. We can query the model and have standardized BMI functions return their names and values (and even their units!). This means no more spending hours poring over code to get the simplest infoâ€”you can use the same functions over and over again to get the info you require.

```python
# we can also look at all values with a loop
input_vars = x.get_input_var_names()
for tmp in input_vars:
    print(tmp, "=", x.get_value_ptr(tmp), x.get_var_units(tmp))
output_vars = x.get_output_var_names()
for tmp in output_vars:
    print(tmp, "=", x.get_value_ptr(tmp), x.get_var_units(tmp))
```

Notice how `land_surface_water__runoff_volume_flux` doesn't equal 0? That's because we set an initial discharge value using the model configuration file before we run the first update step. Now we'll run the model below.

```python
# update for a set number of time steps
run_steps = 30
for i in range(run_steps):
    x.update()
    print("discharge =", x.get_value_ptr("land_surface_water__runoff_volume_flux"))
```

Although it's fun to look at printed data and manually control the loops, let's try running an example simulation now using synthetic precipitation data. First, we'll reinitialize the model and then import the data.

```python
# Reinitialize the model
x.initialize("linearres.yaml")

# Import the example data
forcing = pd.read_csv("data/example_ppt_data.csv")
print(forcing.head(10))
```

We can see in the dataframe above that we have the precipitation we need to run `linearresBMI`.

Now we can run a `linearresBMI` update loop based on the number of entries in the forcing data. Importantly, we'll use BMI::set_value to apply the forcing data to the model and BMI::get_value to access model data for plotting and analysis.

```python
# Make an empty array to store the output data
q_output = np.zeros(forcing.day.size)

# Loop through the data and run snowBMI
for i in range(forcing.day.size):
    precip = np.full(1, forcing.ppt_mm[i])
    x.set_value("atmosphere_water__precipitation_leq-volume_flux", precip)
    x.update()
    q_output[i] = x.get_value_ptr("land_surface_water__runoff_volume_flux")
```

```python
# Plot the simulated discharge
plt.figure(figsize=(10, 7))
plt.plot(q_output,label='Sim. Q')
plt.ylabel('Q (mm/d)')
plt.xlabel('Day of Simulation')
plt.legend()
plt.show()
plt.close()
```

There we have it. Let's end this and finalize our run.

```python
x.finalize()
```
